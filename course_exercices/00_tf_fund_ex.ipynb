```python
import tensorflow as tf
tf.__version__
```

```python
# scalar, matrix, tensor (everything)
scalar = tf.constant(1)
matrix = tf.constant([[1, 2],
                      [3, 4]])
tensor = matrix
```

```python
# shape, rank, size
tensor.shape, tensor.ndim, tf.size(tensor)
```

```python
# reate two tensors containing random values between 0 and 1 with shape [5, 300]
random_tensor1 = tf.random.uniform(minval=0, maxval=1, shape=[5, 300])
random_tensor2 = tf.random.uniform(minval=0, maxval=1, shape=[5, 300])

random_tensor1, random_tensor2
```

```python
# Multiply the two tensors you created in 3 using matrix multiplication.
random_tensor1 * 3, random_tensor2 * 3
```

```python
# Multiply the two tensors you created in 3 using dot product.
random_tensor1 @ 3, random_tensor2 @ 3
```

```python
# Create a tensor with random values between 0 and 1 with shape [224, 224, 3]
tensor3 = tf.random.uniform(shape=[224, 224, 3])
tensor3
```

```python
# Find the min and max values of the tensor you created in 6 along the first axis.
tf.reduce_min(tensor3, 1), tf.reduce_max(tensor3, 1), 
```

```python
# Created a tensor with random values of shape [1, 224, 224, 3] then squeeze it to change the shape to [224, 224, 3].
tensor4 = tf.random.uniform(shape=[1, 224, 224, 3])
tensor4
```
